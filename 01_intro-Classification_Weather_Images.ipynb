### 3. Define label function 

```python
# def is_cat(x): return x[0].isupper()

def is_category(x): 
    if 'cloudy' in x: return 'cloudy'
    elif 'rain' in x: return 'rain'
    elif 'shine' in x: return 'shine'
    elif 'sunrise' in x: return 'sunrise'
```

```python
path_imgs.ls()[0],is_category(str(path_imgs.ls()[0]))
```

### 4. Define outputs folder

```python
# create path to save outputs
path_outputs = path/'outputs'
path_outputs
```

### 5. Define dataloaders (training + validation)

```python
%%time

# dls = ImageDataLoaders.from_name_func(
#     path, get_image_files(path_to_images), valid_pct=0.2, seed=42,
#    label_func=is_cat, item_tfms=Resize(224))


dls = ImageDataLoaders.from_name_func(
    path_outputs, 
    get_image_files(path_imgs), 
    valid_pct=0.2, seed=42,
    label_func=is_category, 
    item_tfms=Resize(224),
    # bs=64
)
```

```python
# batch size
dls.bs
```

```python
# categories
dls.vocab
```

```python
# Exemples of training images 
dls.train.show_batch(max_n=4, nrows=1)
```

```python
# Exemples of validation images 
dls.valid.show_batch(max_n=4, nrows=1)
```

### 6. Define learner

```python
model = resnet34
# model = resnet50
learn = cnn_learner(dls, model, metrics=error_rate)
```

```python
learn.path
```

### 7. Training

```python
# resnet34
learn.fine_tune(1)
```

```python
learn.save("weather_resnet34")
```

```python
# resnet50
# learn.fine_tune(1)
```

A bigger model (resnet50 > resnet34) does not always give a better performance because it needs more data in particular.

**Advice**: start training with a small model for basic performance before testing larger models.

```python
# save model
learn.save("weather_resnet50")
```

```python
# load saved model
learn.load("weather_resnet34")
```

```python
# Confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
# Worst predictions
interp.plot_top_losses(5, nrows=1)
```
